```python
import torch
from diffusers import StableDiffusionPipeline

pipe = StableDiffusionPipeline.from_pretrained("CompVis/stable-diffusion-v1-4", torch_dtype=torch.float16)
pipe = pipe.to("cuda")
```

```python
def text_to_image(prompt):
  generator = torch.Generator("cuda").manual_seed(2048)

  image = pipe(prompt, num_inference_steps=70, generator=generator).images[0]

  return image
```

```python
from flask import Flask, request, jsonify
import torch
import io
from PIL import Image, ImageOps
import numpy as np
import tempfile
import os
import gc
import base64
import json
import uuid
from pyngrok import ngrok
from IPython.display import display
import subprocess
```

```python
def get_backgroundless_image(image_name):
    directory = '/content/backgroundless-images/'
    image_path = os.path.join(directory, image_name)
    if not os.path.exists(image_path):
        raise FileNotFoundError(f"No image found with the name: {image_name}")
    image = Image.open(image_path)
    display(image)
    return image_path
```

```python
def remove_background(image_path):
    output_directory = '/content/backgroundless-images/'
    os.makedirs(output_directory, exist_ok=True)

    !removebg --api-key $RMBG_API_KEY --output-directory $output_directory $image_path
    output_file_name = image_path.split('images/')[-1]
    print(output_file_name)
    backgroundless_image_path = get_backgroundless_image(output_file_name)

    return backgroundless_image_path
```

```python
def generate_unique_filename(extension=".png"):
    unique_id = uuid.uuid4()
    filename = f"{unique_id}{extension}"
    return filename
```

```python
def save_image(image):
  # Convert PIL Image to BytesIO object
  img_byte_arr = io.BytesIO()
  image.save(img_byte_arr, format='PNG')
  img_byte_arr = img_byte_arr.getvalue()
  # Convert BytesIO object back to PIL Image
  image = Image.open(io.BytesIO(img_byte_arr))
  folder_path = '/content/images/'
  image_name = generate_unique_filename('.png')
  image_path = f'{folder_path}{image_name}'
  os.makedirs(os.path.dirname(folder_path), exist_ok=True)
  image.save(image_path)
  return image_path
```

```python
def encode_files(image_path):
  with open(image_path, 'rb') as f:
    image = f.read()

  encoded_image = base64.b64encode(image).decode('utf-8')

  return encoded_image
```

```python
def establish_connection():
  port = 5000
  public_url = ngrok.connect(port).public_url
  print(' * ngrok tunnel "{}" -> "http://127.0.0.1:{}"'.format(public_url, port))
  app.run(host='0.0.0.0', port=port)
```

```python
def validate_request_prompt(request):
    try:
        json_data = request.data.decode('utf-8')
        data = json.loads(json_data)
        prompt = data.get('prompt', '')
    except Exception as e:
        print(f"Error decoding request data: {e}")
        prompt = ''
        return None, False
    return prompt, True
```

```python
app = Flask(__name__)

@app.route('/generate_from_text', methods=['POST'])
def generate_from_text():
    prompt, valid = validate_request_prompt(request)
    if valid:
      image = text_to_image(prompt)
      display(image)
      if image:
          image_path = save_image(image)

          backgroundless_image_path = remove_background(image_path)

          encoded_image = encode_files(backgroundless_image_path)

          response_data = {
              'image': encoded_image
          }

          gc.collect()
          torch.cuda.empty_cache()

          return jsonify(response_data)

@app.route('/hi', methods=['GET'])
def hi():
  return jsonify({'message': 'Hello, World!'})

if __name__ == '__main__':
  establish_connection()
```
